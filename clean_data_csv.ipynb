```python
import pandas as pd
```

```python
df = pd.read_csv('jobs.db/academy_awards.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df['Won?'].value_counts()
```

```python
accepted_values_won=["YES","NO"]
```

```python
df=df[df['Won?'].isin(accepted_values_won)]
```

```python
df['Won?'].value_counts()
```

```python
df['Year'] = df['Year'].str[0:4]
```

```python
df.head()
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['Year'].dtypes
```

```python
df_later_than_2000=df[df['Year']>2000]
```

```python
df_later_than_2000
```

### Nettoyer la colonne Won?

```python
replacements = {"NO": 0, "YES": 1}
```

```python
df = df_later_than_2000
```

```python
df['Won?']
```

```python
df['Won?'] = df['Won?'].map(replacements)
```

```python
print(df['Won?'])
```

```python
df['Won?'].dtypes
```

```python
df.head()
```

```python
df['Won']=df['Won?']
```

```python
df.head()
```

```python
drop_cols=['Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10','Won?']
```

```python
final_nominations = df.drop(drop_cols, axis=1)
```

```python
final_nominations.head()
```

```python
movie = df['Additional Info'].str.rstrip("'}").str.split(" {'", expand=True)[0]
```

```python
movie.head()
```

```python
character = df['Additional Info'].str.rstrip("'}").str.split(" {'", expand=True)[1]
```

```python
character.head()
```

```python
final_nominations['movie']=movie
final_nominations['character']=character
```

```python
final_nominations.head()
```

```python
final_nominations.drop('Additional Info', axis=1)
```

### Exporter vers sqlite
